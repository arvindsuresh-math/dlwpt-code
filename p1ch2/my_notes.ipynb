```python
# This module provides a set of functions for image transformations using torchvision.
#
# The `transforms` module contains various classes and functions to apply transformations to images, such as resizing, cropping, flipping, and more.
#
# Example usage:
#     transform = transforms.Compose([
#         transforms.Resize(256),
#         transforms.CenterCrop(224),
#         transforms.ToTensor(),
#     ])

from torchvision import transforms
```

```python

preprocess = transforms.Compose([
                transforms.Resize(256),
                transforms.CenterCrop(224),
                transforms.ToTensor(),
                transforms.Normalize(
                                    mean=[0.485, 0.456, 0.406],
                                    std=[0.229, 0.224, 0.225]
                                    )])

"""
Preprocesses an image using a series of transformations.

Parameters:
    - resize (int): The size to which the image should be resized.
    - crop (int): The size of the center crop.
    - mean (list): The mean values for normalization.
    - std (list): The standard deviation values for normalization.

Returns:/

    - tensor: The preprocessed image as a tensor.
"""
```

```python
from PIL import Image
```

```python
#create image objects with the two images anbu.jpg and murphy.jpg
anbu = Image.open("../data/p1ch2/anbu.jpeg")
murphy = Image.open("../data/p1ch2/murphy.jpeg")
```

```python
anbu
```

```python
murphy
```

```python
anbu_preprocessed = preprocess(anbu)
```

```python
anbu_preprocessed
```

```python
murphy_preprocessed = preprocess(murphy)
```

```python
murphy_preprocessed
```

```python
batch_anbu = torch.unsqueeze(anbu_preprocessed, 0)
```

```python
batch_murphy = torch.unsqueeze(murphy_preprocessed, 0)
```

```python
resnet.eval()
```

```python
out_anbu = resnet(batch_anbu)
out_murphy = resnet(batch_murphy)
```

```python
with open('../data/p1ch2/imagenet_classes.txt') as f:
    labels = [line.strip() for line in f.readlines()]
```

```python
_, index_anbu = torch.max(out_anbu, 1)
_, index_murphy = torch.max(out_murphy, 1)
```

```python
percentage_anbu = torch.nn.functional.softmax(out_anbu, dim=1)[0] * 100
print(labels[index_anbu[0]], percentage_anbu[index_anbu[0]].item())
```

```python
percentage_murphy = torch.nn.functional.softmax(out_murphy, dim=1)[0] * 100
print(labels[index_murphy[0]], percentage_murphy[index_murphy[0]].item())
```

```python
_,indices_anbu = torch.sort(out_anbu, descending=True)
[(labels[idx], percentage_anbu[idx].item()) for idx in indices_anbu[0][:5]]
```

```python
_,indices_murphy = torch.sort(out_murphy, descending=True)
[(labels[idx], percentage_murphy[idx].item()) for idx in indices_murphy[0][:5]]
```
